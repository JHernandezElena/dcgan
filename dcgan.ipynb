### Import packages

```python
from dcgan import DCGAN
```

### Train

```python
dcgan = DCGAN('fashion_mnist') #cargar dos tipos de imagenes (digitos manuscritos-mnist o prendas de ropa)
dcgan.train(epochs=4000, batch_size=32, save_interval=50)

    #se puede conseguir una red que creee imagenes creibles de digitos y/o prendas
    #con 1000 epochs ya se pueden medio ver digits, con 2000 mucho mejor y con 4000 un poco mejor
        #con prendas se nota mucho mucho mejor (de 2000 a 4000 no hay mucha diferencia)
```
